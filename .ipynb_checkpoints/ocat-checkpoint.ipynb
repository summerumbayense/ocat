```julia
using SymPy
using CSV
```

```julia
datos = CSV.read("entrenamiento.data", header = false)
```

```julia
ent = Bool[]
for i = 1:5716
    length(unique(datos[:,i])) >= 2 ? push!(ent, true) : push!(ent, false)
end
datos = datos[ent]
datos = datos[:,union(1:2000, 3593)]
```

```julia
negativos = datos[:,size(datos,2)] .!= 3
negativos = filter(i -> negativos[i], 1:size(datos, 1))
positivos = setdiff(1:size(datos, 1), negativos)
```

```julia
function posNeg(columna::Symbol)
    pos = datos[positivos,:][columna]
    pos = filter(i -> pos[i] == 1, 1:size(pos, 1))
    
    neg = datos[negativos,:][columna]
    neg = filter(i -> neg[i] == 1, 1:size(neg, 1))
    pos, neg
end
```

```julia
function maximo(columnas::Array{Symbol,1}, pos_ex::Array{Int64,1}, neg_ex::Array{Int64,1})
    aptitudes = Float64[]
    valores = Tuple{Array{Int64,1},Array{Int64,1}}[]
    negex = setdiff(1:length(negativos), neg_ex)
    for columna = columnas
        pos, neg = posNeg(columna)
        p = setdiff(pos, pos_ex)
        n = setdiff(neg, negex)
        if ~isempty(p)
            if length(n) > 0
                push!(aptitudes, length(p)/length(n))
                push!(valores, (p, n))
            else
                push!(aptitudes, length(p)+1000)
                push!(valores, (p, n))
            end
        else
            n = setdiff(negativos, n)
            push!(aptitudes, -length(n))
            push!(valores, (p, n))
        end
    end
    maximo = argmax(aptitudes)
    columnas[maximo], valores[maximo]
end        
```

```julia
function ocat()
    neg = 1:length(negativos)
    pos = 1:length(positivos)
    clausulas = []
    neg_ex = Array(1:length(negativos))
    
    while ~isempty(neg)
        c = false
        nuevos_neg = Int64[]
        columnas = names(datos)
        pos_ex = Int64[]
        
        while ~isempty(pos)
            if isempty(neg_ex)
                maxi, vals = maximo(columnas, pos_ex, Array(1:length(negativos)))
            else
                maxi, vals = maximo(columnas, pos_ex, neg_ex)
            end
            println(maxi)
            npos, nneg = vals[1], vals[2]
            pos_ex = union(pos_ex, npos)
            println(npos)
            if ~isempty(npos)
                c == false ? c = (c || Sym(String(maxi))) : c = Or(c, Sym(String(maxi)))
            else
                c = Or(c, Not(Sym(String(maxi))))
                append!(nuevos_neg, nneg)
                break
            end
            pos = setdiff(pos, npos)
            println(length(pos))
            append!(nuevos_neg, nneg)
            setdiff!(columnas, [maxi])
        end
        
        push!(clausulas, c)
        pos = 1:length(positivos)
        intersect!(neg_ex, nuevos_neg)
        println(length(neg_ex))
        neg = copy(neg_ex)
    end
    
    clausulas
end            
```
